```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
#%matplotlib inline  
```

```python
# Configurations
# 1 client VM
# CT=2
wt = [8, 16, 32, 64] 
vc = [1, 4, 8, 16, 24, 32, 40]
figure_path = "/home/simon/eth/asl-fall17-project/report/Data/31/"
```

```python
# 3.1 One Middleware
mw_gets_tp = {}
mw_gets_rt = {}
mw_sets_tp = {}
mw_sets_rt = {}

max_get_tps = []
max_set_tps = []
min_get_rts = []
min_set_rts = []

# Minimum service time found
S_server_get = 1000
S_server_set = 1000

data_path = "/home/simon/eth/asl-fall17-project/experiment-scripts/3_Baseline_w_MW/31_one_middleware/logs/"
for threads in wt:
    max_get_tp = 0
    max_set_tp = 0
    min_get_rt = 10000
    min_get_st = 10000
    min_set_rt = 10000
    min_set_st = 10000
    
    for clients in vc:
        # GETS
        path = data_path+"/wt"+str(threads)+"/VC_"+str(clients)+"/GETS/"
        for directory, subdirectory, files in os.walk(path):
            if(len(files) == 4):
                # We're in a MW folder
                table = pd.read_csv(directory+"/MeasurementTable.csv",index_col=False)
                rt_histogram = pd.read_csv(directory+"/ResponseTimeHistogram.csv")
                throughput = np.sum(table['get'][1:])/np.max(table['time'])/threads

                for i in range(1,len(table['get'])):
                    max_get_tp = np.maximum(max_get_tp, table['get'][i]/(table['time'][i]-table['time'][i-1]))
                
                resptime = 0
                total_count = 0
                
                for row in range(rt_histogram.shape[0]):
                    total_count += rt_histogram.iloc[row,:]['requestcount']
                    resptime += rt_histogram.iloc[row,:]['requestcount']*rt_histogram.iloc[row,:]['time']
                    if(rt_histogram.iloc[row,:]['requestcount'] > 0):
                        min_get_st = np.minimum(min_get_st,rt_histogram.iloc[row,:]['time'])
                # Convert to ms from microseconds
                resptime = resptime/total_count/1000.0
                
                key = str("wt"+str(threads)+"vc"+str(clients))
                if key not in mw_gets_tp:
                    mw_gets_tp[key] = []
                    mw_gets_rt[key] = []
                
                mw_gets_tp[key].append(throughput)
                mw_gets_rt[key].append(resptime)
                
                S_server_get = np.minimum(S_server_get, np.min(table['avgGetServiceTime']))
                
                
        # SETS
        path = data_path+"/wt"+str(threads)+"/VC_"+str(clients)+"/SETS/"
        for directory, subdirectory, files in os.walk(path):
            if(len(files) == 4):
                # We're in a MW folder
                table = pd.read_csv(directory+"/MeasurementTable.csv",index_col=False)
                rt_histogram = pd.read_csv(directory+"/ResponseTimeHistogram.csv")
                throughput = np.sum(table['set'][1:])/np.max(table['time'])/threads
                for i in range(1,len(table['set'])):
                    max_set_tp = np.maximum(max_set_tp, table['set'][i]/(table['time'][i]-table['time'][i-1]))
                resptime = 0
                total_count = 0
                
                for row in range(rt_histogram.shape[0]):
                    total_count += rt_histogram.iloc[row,:]['requestcount']
                    resptime += rt_histogram.iloc[row,:]['requestcount']*rt_histogram.iloc[row,:]['time']
                    if(rt_histogram.iloc[row,:]['requestcount'] > 0):
                        min_set_st = np.minimum(min_set_st,rt_histogram.iloc[row,:]['time'])
                # Convert to ms from microseconds
                resptime = resptime/total_count/1000.0
                
                key = str("wt"+str(threads)+"vc"+str(clients))
                if key not in mw_sets_tp:
                    mw_sets_tp[key] = []
                    mw_sets_rt[key] = []

                mw_sets_tp[key].append(throughput)
                mw_sets_rt[key].append(resptime)
                
                S_server_set = np.minimum(S_server_set, np.min(table['avgSetServiceTime']))
    max_get_tps.append(max_get_tp)
    max_set_tps.append(max_set_tp)
    min_get_rts.append(min_get_st)
    min_set_rts.append(min_set_st)
                
#8, 16, 32, 64
get_mean_tp = []
get_stddev_tp = []
get_mean_rt = []
get_stddev_rt = []
set_mean_tp = []
set_stddev_tp = []
set_mean_rt = []
set_stddev_rt = []

for threads in wt:
    g_m_tp = []
    g_s_tp = []
    g_m_rt = []
    g_s_rt = []

    s_m_tp = []
    s_s_tp = []
    s_m_rt = []
    s_s_rt = []

    for clients in vc:
        key = str("wt"+str(threads)+"vc"+str(clients))
        #g_m_tp.append(np.mean(mw_gets_tp[key]))
        g_m_tp.append(np.mean(mw_gets_tp[key]))
        g_s_tp.append(np.std(mw_gets_tp[key]))
        g_m_rt.append(np.mean(mw_gets_rt[key]))
        g_s_rt.append(np.std(mw_gets_rt[key]))
        
        #s_m_tp.append(np.mean(mw_sets_tp[key]))
        s_m_tp.append(np.mean(mw_sets_tp[key]))
        s_s_tp.append(np.std(mw_sets_tp[key]))
        s_m_rt.append(np.mean(mw_sets_rt[key]))
        s_s_rt.append(np.std(mw_sets_rt[key]))
        
    
    get_mean_tp.append(g_m_tp)
    get_stddev_tp.append(g_s_tp)
    get_mean_rt.append(g_m_rt)
    get_stddev_rt.append(g_s_rt)
    set_mean_tp.append(s_m_tp)
    set_stddev_tp.append(s_s_tp)
    set_mean_rt.append(s_m_rt)
    set_stddev_rt.append(s_s_rt)

print("S_server_get: " +str(S_server_get))
print("S_server_set: " +str(S_server_set))
print("Maximum get TPs: " + str(max_get_tps))
print("Maximum set TPs: " + str(max_set_tps))
print("Min get RTs: " + str(min_get_rts))
print("Min set RTs: " + str(min_set_rts))
print (str(get_mean_rt))
```

```python
# 3.2 Two Middlewares
mw1_gets_tp = {}
mw1_gets_rt = {}
mw1_sets_tp = {}
mw1_sets_rt = {}
mw2_gets_tp = {}
mw2_gets_rt = {}
mw2_sets_tp = {}
mw2_sets_rt = {}

max_get_tps2 = []
max_set_tps2 = []
min_get_rts = []
min_set_rts = []

data_path = "/home/simon/eth/asl-fall17-project/experiment-scripts/3_Baseline_w_MW/32_two_middlewares/logs/"
for threads in wt:
    max_get_tp = 0
    max_set_tp = 0
    min_get_rt = 100000
    min_set_rt = 100000
    for clients in vc:
        # GETS
        path = data_path+"/wt"+str(threads)+"/VC_"+str(clients)+"/GETS/"
        for directory, subdirectory, files in os.walk(path):
            if(subdirectory == ['MW2', 'MW1']):
                # MW1
                for directory, subdirectory, files in os.walk(path+"MW1/"):
                    if(len(files) ==4 ):
                        # We're in a MW folder
                        table = pd.read_csv(directory+"/MeasurementTable.csv",index_col=False)
                        rt_histogram = pd.read_csv(directory+"/ResponseTimeHistogram.csv")
                        throughput = np.sum(table['get'][1:])/np.max(table['time'])/threads
                        for i in range(1,len(table['get'])):
                            max_get_tp = np.maximum(max_get_tp, table['get'][i]/(table['time'][i]-table['time'][i-1]))
                        resptime = 0
                        total_count = 0

                        for row in range(rt_histogram.shape[0]):
                            total_count += rt_histogram.iloc[row,:]['requestcount']
                            resptime += rt_histogram.iloc[row,:]['requestcount']*rt_histogram.iloc[row,:]['time'] 
                            if(rt_histogram.iloc[row,:]['requestcount'] > 0):
                                min_get_rt = np.minimum(min_get_rt,rt_histogram.iloc[row,:]['time'])
                        # Convert to ms from microseconds
                        resptime = resptime/total_count/1000.0

                        key = str("wt"+str(threads)+"vc"+str(clients))
                        if key not in mw1_gets_tp:
                            mw1_gets_tp[key] = []
                            mw1_gets_rt[key] = []

                        mw1_gets_tp[key].append(throughput)
                        mw1_gets_rt[key].append(resptime)
                        
                        S_server_get = np.minimum(S_server_get, np.min(table['avgGetServiceTime']))
#                         T_max_get = np.maximum(T_max_get, np.max(throughput))
                
                
                # MW2
                for directory, subdirectory, files in os.walk(path+"MW2/"):
                    if(len(files) ==4 ):
                        # We're in a MW folder
                        table = pd.read_csv(directory+"/MeasurementTable.csv",index_col=False)
                        rt_histogram = pd.read_csv(directory+"/ResponseTimeHistogram.csv")
                        throughput = np.sum(table['get'][1:])/np.max(table['time'])/threads
                        for i in range(1,len(table['get'])):
                            max_get_tp = np.maximum(max_get_tp, table['get'][i]/(table['time'][i]-table['time'][i-1]))
                        resptime = 0
                        total_count = 0

                        for row in range(rt_histogram.shape[0]):
                            total_count += rt_histogram.iloc[row,:]['requestcount']
                            resptime += rt_histogram.iloc[row,:]['requestcount']*rt_histogram.iloc[row,:]['time'] 
                            if(rt_histogram.iloc[row,:]['requestcount'] > 0):
                                min_get_rt = np.minimum(min_get_rt,rt_histogram.iloc[row,:]['time'])
                        # Convert to ms from microseconds
                        resptime = resptime/total_count/1000.0

                        key = str("wt"+str(threads)+"vc"+str(clients))
                        if key not in mw2_gets_tp:
                            mw2_gets_tp[key] = []
                            mw2_gets_rt[key] = []

                        mw2_gets_tp[key].append(throughput)
                        mw2_gets_rt[key].append(resptime)
                        
                        S_server_get = np.minimum(S_server_get, np.min(table['avgGetServiceTime']))
#                         T_max_get = np.maximum(T_max_get, np.max(throughput))
                
                
        # SETS
        path = data_path+"/wt"+str(threads)+"/VC_"+str(clients)+"/SETS/"
        for directory, subdirectory, files in os.walk(path):
            if(subdirectory == ['MW2', 'MW1']):
                # MW1
                for directory, subdirectory, files in os.walk(path+"MW1/"):
                    if(len(files) == 4 ):
                        # We're in a MW folder
                        table = pd.read_csv(directory+"/MeasurementTable.csv",index_col=False)
                        rt_histogram = pd.read_csv(directory+"/ResponseTimeHistogram.csv")
                        throughput = np.sum(table['set'][1:])/np.max(table['time'])/threads
                        for i in range(1,len(table['set'])):
                            max_set_tp = np.maximum(max_set_tp, table['set'][i]/(table['time'][i]-table['time'][i-1]))
                        
                        resptime = 0
                        total_count = 0

                        for row in range(rt_histogram.shape[0]):
                            total_count += rt_histogram.iloc[row,:]['requestcount']
                            resptime += rt_histogram.iloc[row,:]['requestcount']*rt_histogram.iloc[row,:]['time']
                            if(rt_histogram.iloc[row,:]['requestcount'] > 0):
                                min_set_rt = np.minimum(min_set_rt,rt_histogram.iloc[row,:]['time'])
                        # Convert to ms from microseconds
                        resptime = resptime/total_count/1000.0

                        key = str("wt"+str(threads)+"vc"+str(clients))
                        if key not in mw1_sets_tp:
                            mw1_sets_tp[key] = []
                            mw1_sets_rt[key] = []

                        mw1_sets_tp[key].append(throughput)
                        mw1_sets_rt[key].append(resptime)
                        
                        S_server_set = np.minimum(S_server_set, np.min(table['avgSetServiceTime']))
#                         T_max_set = np.maximum(T_max_set, np.max(throughput))
                
                
                # MW2
                for directory, subdirectory, files in os.walk(path+"MW2/"):
                    if(len(files) ==4 ):
                        # We're in a MW folder
                        table = pd.read_csv(directory+"/MeasurementTable.csv",index_col=False)
                        rt_histogram = pd.read_csv(directory+"/ResponseTimeHistogram.csv")
                        throughput = np.sum(table['set'][1:])/np.max(table['time'])/threads
                        for i in range(1,len(table['set'])):
                            max_set_tp = np.maximum(max_set_tp, table['set'][i]/(table['time'][i]-table['time'][i-1]))
                        
                        resptime = 0
                        total_count = 0

                        for row in range(rt_histogram.shape[0]):
                            total_count += rt_histogram.iloc[row,:]['requestcount']
                            resptime += rt_histogram.iloc[row,:]['requestcount']*rt_histogram.iloc[row,:]['time']
                            if(rt_histogram.iloc[row,:]['requestcount'] > 0):
                                min_set_rt = np.minimum(min_set_rt,rt_histogram.iloc[row,:]['time'])
                        # Convert to ms from microseconds
                        resptime = resptime/total_count/1000.0

                        key = str("wt"+str(threads)+"vc"+str(clients))
                        if key not in mw2_sets_tp:
                            mw2_sets_tp[key] = []
                            mw2_sets_rt[key] = []

                        mw2_sets_tp[key].append(throughput)
                        mw2_sets_rt[key].append(resptime)
                        
                        S_server_set = np.minimum(S_server_set, np.min(table['avgSetServiceTime']))
#                         T_max_set = np.maximum(T_max_set, np.max(throughput))
    max_get_tps2.append(max_get_tp)
    max_set_tps2.append(max_set_tp)
    min_get_rts.append(min_get_rt)
    min_set_rts.append(min_set_rt)
        
        
        
# Gather Aggregates
#8, 16, 32, 64
get_mean_tp = []
get_stddev_tp = []
get_mean_rt = []
get_stddev_rt = []
set_mean_tp = []
set_stddev_tp = []
set_mean_rt = []
set_stddev_rt = []

for threads in wt:
    g_m_tp = []
    g_s_tp = []
    g_m_rt = []
    g_s_rt = []

    s_m_tp = []
    s_s_tp = []
    s_m_rt = []
    s_s_rt = []
    for clients in vc:
        key = str("wt"+str(threads)+"vc"+str(clients))
        
        # Handle incomplete repetitions
        if not len(mw1_gets_tp[key]) == len(mw2_gets_tp[key]):
            length = np.min([len(mw1_gets_tp[key]), len(mw2_gets_tp[key])])
            mw1_gets_tp[key] = mw1_gets_tp[key][:length-1]
            mw2_gets_tp[key] = mw2_gets_tp[key][:length-1]

        
#         print("mw1")
#         print(mw1_gets_tp[key])
#         print("mw2")
#         print(mw2_gets_tp[key])
#         print("Added")
#         print(np.add(mw1_gets_tp[key],mw2_gets_tp[key]))
#         print("mean")
#         print(np.mean(np.add(mw1_gets_tp[key],mw2_gets_tp[key])))
        g_m_tp.append(np.mean(np.add(mw1_gets_tp[key],mw2_gets_tp[key])))
        g_s_tp.append(np.std(np.add(mw1_gets_tp[key],mw2_gets_tp[key])))
        g_m_rt.append(np.mean(mw1_gets_rt[key]+mw2_gets_rt[key]))
        g_s_rt.append(np.std(mw1_gets_rt[key]+mw2_gets_rt[key]))
        
        s_m_tp.append(np.mean(np.add(mw1_sets_tp[key],mw2_sets_tp[key])))
        s_s_tp.append(np.std(np.add(mw1_sets_tp[key],mw2_sets_tp[key])))
        s_m_rt.append(np.mean(mw1_sets_rt[key]+mw2_sets_rt[key]))
        s_s_rt.append(np.std(mw1_sets_rt[key]+mw2_sets_rt[key]))
    
    get_mean_tp.append(g_m_tp)
    get_stddev_tp.append(g_s_tp)
    get_mean_rt.append(g_m_rt)
    get_stddev_rt.append(g_s_rt)
    set_mean_tp.append(s_m_tp)
    set_stddev_tp.append(s_s_tp)
    set_mean_rt.append(s_m_rt)
    set_stddev_rt.append(s_s_rt)
print("S_server_get: " +str(S_server_get))
print("S_server_set: " +str(S_server_set))
print("Maximum get TPs in 3.1: " + str(max_get_tps))
print("Maximum set TPs in 3.1: " + str(max_set_tps))
print("Maximum get TPs in 3.2: " + str(max_get_tps2))
print("Maximum set TPs in 3.2: " + str(max_set_tps2))

print("Min get RTs: " + str(min_get_rts))
print("Min set RTs: " + str(min_set_rts))
```
